5. What are iterators in Python and how do they differ from iterables?
- Iterable
- An object **that can be looped over** (e.g., lists, tuples, strings).
- Has a built-in `__iter__()` method.
- Does **not** remember its state (each loop starts fresh).

 **Iterator**
- An object **used to iterate** through an iterable.
- Implements both `__iter__()` **and** `__next__()` methods.
- **Remembers its position**, so each call to `next()` gives the next item.

```python
my_list = [1, 2, 3]   # This is an iterable
for item in my_list:
    print(item)

my_list = [1, 2, 3]
it = iter(my_list)    # Now it's an iterator
print(next(it))       # 1
print(next(it))       # 2
```

6. Explain the concept of generators in Python and how they are defined.
-  A generator is a special type of function that lets you generate values one by one, instead of returning them all at once.

```python
def my_gen():
    yield 1
    yield 2
    yield 3

g = my_gen()

print(next(g))
print(next(g))
print(next(g))
```

```python
#Write a Python function that checks if a given number is prime or not from 1 to 200.
def is_prime(n):
    if n < 2:
        return False
    return all(n % i != 0 for i in range(2, int(n**0.5) + 1))

num_list = set(range(1, 201))
prime_flags = set(map(lambda x: (x, is_prime(x)), num_list))
primes = [x for x, is_p in prime_flags if is_p]
print(primes)
```

```python
#Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.
def fibonacci(n):
  a,b=0,1
  out=[]
  for _ in range(n):
    out.append(a)
    a,b=b,b+a
  return out

print(fibonacci(20))
```

```python
# Write a generator function in Python that yields the powers of 2 up to a given exponent.
def power(enter):
  for i in range(enter + 1):
    yield 2**i

n=int(input())
for x in power(n):
  print(x)


```

```python
#9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.
celsius = [0, 20, 37, 100]

fahrenheit = list(map(lambda c: (c * 9/5) + 32, celsius))

print("Fahrenheit:", fahrenheit)
```

```python
#Create a Python program that uses `filter()` to remove all the vowels from a given string.
def pavan(*a):
  for word in a:
    result = list(filter(lambda x: x.lower() not in "aeiou",word))
    print("".join(result))

pavan("pavan")
```
